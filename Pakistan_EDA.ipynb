# Data Quality Report
## Pakistan's Largest E-Commerce Dataset

### 1. Dataset Overview
This dataset contains e-commerce transaction data from Pakistan's largest 
online shopping platform. It covers orders from March 2016 to August 2018. 
The dataset was sourced from Kaggle 
(https://www.kaggle.com/datasets/zusmani/pakistans-largest-ecommerce-dataset) 
and was collected by Zeeshan Usmani. After loading, the dataset contains 
1,048,575 rows and 26 columns, however only 584,524 rows contain real data.

### 2. Missing Values
Several columns contain significant missing values. The most critical is 
sales_commission_code with 601,229 missing values â more than the total real 
rows in the dataset. category_name_1 has 464,215 missing values. Additionally, 
columns Unnamed: 21 through Unnamed: 25 have zero non-null values meaning they 
are completely empty and serve no purpose in analysis.

### 3. Dead Rows and Duplicate Rows
Out of 1,048,575 total rows loaded, 464,051 rows are completely empty. These 
blank rows were loaded from the bottom of the CSV file. This means 44.2% of 
the loaded data is unusable. The duplicated().sum() function confirmed 464,050 
duplicate rows, which matches the empty row count exactly.

### 4. Outliers
Using the IQR method, significant outliers were found in three numeric columns.
The price column has 105,687 outliers with an upper bound of 9,635 PKR. The 
grand_total column has 77,090 outliers with an upper bound of 16,080 PKR. The 
qty_ordered column has 79,310 outliers â this happened because 75% of orders 
contain exactly 1 item, making the IQR equal to zero and flagging anything 
above 1 as an outlier.

### 5. Suspicious and Incorrect Values
Multiple data quality issues were found beyond missing values. The grand_total 
column contains negative values with a minimum of -1,594 PKR which is 
impossible for an order total. The discount_amount column also contains 
negative values with a minimum of -599 PKR. The status and category_name_1 
columns contain '\N' as text values instead of proper null values â this is a 
database export error. The Year and Month columns are stored as float64 instead 
of integers which is incorrect. Additionally 34.43% of all orders were canceled 
which may indicate customer experience problems.

### 6. Week 2 Plan
In Week 2 the following issues will be addressed. First the 464,051 empty rows 
will be dropped using dropna(). The five completely empty unnamed columns will 
be deleted. The '\N' text values in status and category_name_1 will be replaced 
with proper NaN values using replace(). Negative values in grand_total and 
discount_amount will be investigated and either corrected or removed. The Year 
and Month columns will be converted from float64 to integer using astype(). 
Finally outliers in price and grand_total will be capped at their respective 
upper bounds to prevent them from skewing future analysis.

```python
def clean_data(df):
    """
    Cleans the dataframe by:
    - Dropping columns with >50% missing values
    - Removing duplicate/blank rows
    - Imputing numerical columns with median (robust to outliers, unlike mean)
    - Imputing categorical columns with mode (most frequent value)
    - Converting date columns to datetime
    - Creating 2 derived features
    """
    df = df.copy()
    
    # Step 1: Drop columns with more than 50% missing values
    missing_pct = df.isnull().sum() / len(df) * 100
    cols_to_drop = missing_pct[missing_pct > 50].index.tolist()
    df = df.drop(columns=cols_to_drop)
    print(f"Dropped columns (>50% missing): {cols_to_drop}")
    
    # Step 2: Drop completely empty/blank rows
    before = df.shape[0]
    df = df.dropna(how='all')
    after = df.shape[0]
    print(f"Dropped {before - after} completely empty rows")
    
    # Step 3: Remove duplicate rows
    before = df.shape[0]
    df = df.drop_duplicates()
    after = df.shape[0]
    print(f"Dropped {before - after} duplicate rows")
    
    # Step 4: Impute missing values
    num_cols = df.select_dtypes(include=['float64', 'int64']).columns
    for col in num_cols:
        df[col] = df[col].fillna(df[col].median())
    
    cat_cols = df.select_dtypes(include=['object']).columns
    for col in cat_cols:
        df[col] = df[col].fillna(df[col].mode()[0])
    
    print("Missing values after imputation:", df.isnull().sum().sum())
    
    # Step 5: Convert date columns to datetime
    df['created_at'] = pd.to_datetime(df['created_at'], errors='coerce')
    df['Working Date'] = pd.to_datetime(df['Working Date'], errors='coerce')
    print("Date columns converted to datetime")
    
    # Step 6: Create 2 derived features
    # Feature 1: order_year - extracted from created_at date
    df['order_year'] = df['created_at'].dt.year
    
    # Feature 2: is_high_value - binary flag, 1 if price is above median price, else 0
    median_price = df['price'].median()
    df['is_high_value'] = (df['price'] > median_price).astype(int)
    
    print("Created 2 derived features: order_year, is_high_value")
    
    return df
```

**Insight:** Mobiles & Tablets was the dominant category in 2017 with the highest order volume among the top 5. 
However, Men's Fashion overtook it in 2018, suggesting a possible shift in customer purchasing trends.

# Data Story: Pakistan's E-Commerce Dataset (2016-2018)

## Orders Growing Then Dropping
When I grouped the data by year, I saw orders went from 133,909 in 2016 up to 290,920 in 2017 â that's 
almost double. Then 2018 shows only 159,695 orders, which looks like a big drop. But this is not really a 
drop, because our dataset only has 8 months of data for 2018, not the full year. So the business was 
probably still growing, we just don't have the full picture for that year. I also checked average price 
per month, and found May and February had the highest average prices, while August and December had the 
lowest. I'm not 100% sure why, but my guess is December has more cheap bulk orders because of holiday 
sales.

## Mobiles Are King, Cash Is King
Mobiles & Tablets was the most ordered category overall, and it was the top category in 2017 specifically. 
But in 2018, Men's Fashion took over as the top category instead, so customer behavior seems to be 
changing. When I made the box plot, Mobiles & Tablets also had the biggest price range (cheap phones to 
expensive ones), while Men's Fashion and Superstore had small boxes near the bottom with a lot of outlier 
dots above them â meaning most items there are cheap but a few are very expensive. For payment, almost 
half of all orders (46%) used Cash on Delivery. This makes sense for Pakistan, where people don't fully 
trust online payment yet, even though other options like Payaxis and Easypay were also used by some 
customers.

## The Biggest Problem: Cancelled Orders
This was the most important thing I found. 34.43% of all orders got canceled â more than 1 in every 3 
orders. My pie chart showed this clearly too: completed orders were 40.3% and canceled orders were 34.7%, 
which is really close to each other. If I add refunds and order_refunded on top of that, more than 45% of 
orders never actually became a real sale. I think this is a serious problem for this business and if I was 
running this company, I would want to know WHY so many orders get canceled, especially since most people 
pay with Cash on Delivery (maybe people cancel because they don't want to pay when the delivery arrives).

## How the Numbers Connect
In my heatmap, qty_ordered and grand_total had the strongest connection at 0.75. This makes sense â if you 
order more items, your total bill goes up too, and my scatter plot showed this with the line going upward. 
For discounts, I noticed a lot of orders got literally zero discount no matter the price. But for the 
orders that DID get a discount, the more expensive the item, the bigger the discount tended to be â this 
also showed an upward trend line in my scatter plot. Other things like order_year and is_high_value didn't 
really connect to much else.

## Why I Used Median Instead of Mean
My histogram of prices showed something important â most orders are priced under 1,000 PKR, but there's a 
long tail of expensive orders stretching all the way to over 1 million PKR. This is called a "right-skewed" 
distribution. This is exactly why I used median instead of mean to fill in missing prices in my cleaning 
function. If I used mean, those few extremely expensive orders would have pulled the average up to an 
unrealistic number, and filling missing values with that number would be wrong. Median ignores those 
extreme values and gives a more realistic "typical" price.

## What I Learned
Putting all these charts together, I think this dataset tells the story of a business that grew fast and 
people clearly love buying mobiles and fashion items with cash on delivery. But there's a real problem with 
how many orders get cancelled, and that's probably the first thing this company should try to fix if they 
want to grow more in the future. Building these 10 charts was difficult for me since it was my first time, 
but going through each one slowly helped me actually understand what each chart shows, instead of just 
copying code without knowing why.
